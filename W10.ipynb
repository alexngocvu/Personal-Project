```python
# Import necessary libraries
import sqlite3
import pandas as pd
# Connect to the database (Assuming the database is already created and populated)
conn = sqlite3.connect('AdvancedSQLLab.db')
#Window Functions
#Objective: Calculate the running total and ranking of sales for each product

# Running Total
running_total_query = """
SELECT
 ProductID,
 SaleDate,
 SaleAmount,
 SUM(SaleAmount) OVER (PARTITION BY ProductID ORDER BY SaleDate) AS
RunningTotal
FROM
 Sales;
"""
running_total_df = pd.read_sql_query(running_total_query, conn)
running_total_df


# Ranking
ranking_query = """
SELECT
 ProductID,
 SaleDate,
 SaleAmount,
 RANK() OVER (PARTITION BY ProductID ORDER BY SaleAmount DESC) AS
Rank
FROM
 Sales;
"""
ranking_df = pd.read_sql_query(ranking_query, conn)
ranking_df

#Common Table Expressions (CTEs)
#Objective: Create a CTE to calculate the total sales for each productand then query the CTE to find products with total sales greater than a specific amount.

# CTE
cte_query = """
WITH TotalSales_CTE AS (
 SELECT
 ProductID,
 SUM(SaleAmount) AS TotalSales
 FROM
 Sales
 GROUP BY
 ProductID
)
SELECT
 ProductID,
 TotalSales
FROM
 TotalSales_CTE
WHERE
 TotalSales > 40;
"""
cte_df = pd.read_sql_query(cte_query, conn)
cte_df

# Import necessary libraries
import sqlite3
import pandas as pd
# Connect to the database (Assuming the database is already created and populated)
conn = sqlite3.connect('AdvancedSQLLab.db')
# Close the database connection
conn.close()
```

```python
import sqlite3
import pandas as pd

# Connect to the database
conn = sqlite3.connect('AdvancedSQLLab.db')

# Define the CTE query
cte_query = """
WITH TotalSales_CTE AS (
SELECT
 ProductID,
 SaleDate,
 SUM(SaleAmount) AS TotalSales
FROM
 Sales
GROUP BY
 ProductID, SaleDate
)
SELECT
 ProductID,
 SaleDate,
 TotalSales
FROM
 TotalSales_CTE
ORDER BY
 SaleDate;
"""

# Execute the CTE query and load the result into a Pandas DataFrame
cte_df = pd.read_sql_query(cte_query, conn)

# Close the database connection
conn.close()

# Print or use the cte_df as needed
print(cte_df)
```
